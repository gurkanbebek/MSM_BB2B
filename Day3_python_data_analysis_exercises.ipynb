## Problem 3: Advanced Plotting (Subplots)

### 💡 Mini-Tutorial: Creating Side-by-Side Subplots
Instead of showing charts vertically one after another, you can place multiple charts inside a single figure using `plt.subplots(rows, columns)`.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5)) # 1 row, 2 columns grid

# Plot on the left subplot layout
axes[0].hist(df['col_A'])
axes[0].set_title("Left Plot")

# Plot on the right subplot layout
axes[1].scatter(df['col_A'], df['col_B'])
axes[1].set_title("Right Plot")

plt.tight_layout()
```
