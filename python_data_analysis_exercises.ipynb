## Problem 1: Cleaning Missing Data & Filtering

### 💡 Mini-Tutorial: Handling `NaN` and Filtering Rows
Real-world data is often messy and contains missing values. In Pandas, these missing values are represented as `NaN` (Not a Number).

* **Dropping missing values:** `df.dropna(subset=['column_name'])` removes rows where specific columns have empty data.
* **Filtering data:** You can filter rows using conditional statements inside square brackets:
  ```python
  large_beds = df[df['icu_beds'] > 100]
  ```

### 📝 The Challenge
Using the ICU beds dataset, write a script to:
1. Drop any rows where `icu_beds` or `hospitals` contains missing (`NaN`) values.
2. Filter the dataset to only include MMSAs that have **more than 5 hospitals**.
3. Print the total number of regions (rows) that met this criteria, and display their top 5 rows.

````python
import pandas as pd

data_url = "https://raw.githubusercontent.com/fivethirtyeight/data/refs/heads/master/covid-geography/mmsa-icu-beds.csv"

# --- YOUR CODE HERE ---
````

### 🔍 Solution (Problem 1)

````python
# Load the data
df_icu_beds = pd.read_csv(data_url)
print(f"Original dataset shape: {df_icu_beds.shape}")

# Drop rows where 'icu_beds' or 'hospitals' contains NaN values
df_clean = df_icu_beds.dropna(subset=['icu_beds', 'hospitals'])
print(f"Shape after dropping NaNs: {df_clean.shape}")

# Filter the dataset for MMSAs with MORE THAN 5 hospitals
df_filtered = df_clean[df_clean['hospitals'] > 5]

# Print results
print(f"Number of regions with > 5 hospitals: {len(df_filtered)}")
print("\nTop 5 rows of filtered data:")
df_filtered.head()
````

## Problem 2: Data Transformation & Custom Column Creation

### 💡 Mini-Tutorial: Modifying Data Types and Creating New Columns
Sometimes columns look like numbers but are stored as text (strings) because of formatting (like `%` symbols). We cannot perform mathematical operations on text!

To clean the text and transform it into numeric data:
```python
# Remove characters and change data type
df['new_col'] = df['old_col'].str.replace('%', '').astype(float)
```

### 📝 The Challenge
Let's figure out the actual number of high-risk people across these locations:
1. Load the dataset again.
2. Convert the `total_percent_at_risk` column into a usable float by removing the `%` sign and dividing by 100 (e.g., `52.88%` becomes `0.5288`).
3. Create a brand new column called `calculated_high_risk_population` by multiplying `total_at_risk` by your new decimal percentage.
4. Calculate and print the `.mean()` of this new column.

````python
import pandas as pd

data_url = "https://raw.githubusercontent.com/fivethirtyeight/data/refs/heads/master/covid-geography/mmsa-icu-beds.csv"

# --- YOUR CODE HERE ---
````

### 🔍 Solution (Problem 2)

````python
df_icu_beds = pd.read_csv(data_url)

# Clean and convert percentage string to float decimal
df_icu_beds['percent_at_risk_decimal'] = df_icu_beds['total_percent_at_risk'].str.replace('%', '').astype(float) / 100

# Create custom column
df_icu_beds['calculated_high_risk_population'] = df_icu_beds['total_at_risk'] * df_icu_beds['percent_at_risk_decimal']

# Compute average
mean_high_risk = df_icu_beds['calculated_high_risk_population'].mean()

print("Sample of transformed data:")
print(df_icu_beds[['MMSA', 'total_percent_at_risk', 'calculated_high_risk_population']].head())

print(f"\nThe average calculated high-risk population across all MMSAs is: {mean_high_risk:,.2f}")
````

## Problem 3: Advanced Plotting (Subplots)

### 💡 Mini-Tutorial: Creating Side-by-Side Subplots
Instead of showing charts vertically one after another, you can place multiple charts inside a single figure using `plt.subplots(rows, columns)`.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5)) # 1 row, 2 columns grid

# Plot on the left subplot layout
axes[0].hist(df['col_A'])
axes[0].set_title("Left Plot")

# Plot on the right subplot layout
axes[1].scatter(df['col_A'], df['col_B'])
axes[1].set_title("Right Plot")

plt.tight_layout()
```

### 📝 The Challenge
Build a 1-row, 2-column subplot figure to compare resources across the MMSAs:
1. **Plot 1 (Left Subplot):** Create a **histogram** of the `icu_beds` column. Customize it with a green color and clear gridlines.
2. **Plot 2 (Right Subplot):** Create a **scatterplot** showing the relationship between `hospitals` (X-axis) and `icu_beds` (Y-axis).
3. Add a global title to the entire figure layout using `plt.suptitle("MMSA Healthcare Capacity Overview")`.

````python
import pandas as pd
import matplotlib.pyplot as plt

data_url = "https://raw.githubusercontent.com/fivethirtyeight/data/refs/heads/master/covid-geography/mmsa-icu-beds.csv"
df_icu_beds = pd.read_csv(data_url).dropna(subset=['icu_beds', 'hospitals'])

# --- YOUR CODE HERE ---
````

### 🔍 Solution (Problem 3)

````python
# Initialize the side-by-side subplot structure (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# --- Plot 1: Histogram (Left Axis - axes[0]) ---
axes[0].hist(df_icu_beds['icu_beds'], bins=20, edgecolor="black", color="forestgreen")
axes[0].set_title("Distribution of ICU Beds")
axes[0].set_xlabel("ICU Beds per 100k People")
axes[0].set_ylabel("Number of MMSAs")
axes[0].grid(axis="y", alpha=0.4)

# --- Plot 2: Scatterplot (Right Axis - axes[1]) ---
axes[1].scatter(df_icu_beds['hospitals'], df_icu_beds['icu_beds'], color="darkblue", alpha=0.7)
axes[1].set_title("Hospitals vs. ICU Beds")
axes[1].set_xlabel("Hospitals per 100k People")
axes[1].set_ylabel("ICU Beds per 100k People")
axes[1].grid(True, alpha=0.3)

# Add global details and adjust spacing
plt.suptitle("MMSA Healthcare Capacity Overview", fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````
